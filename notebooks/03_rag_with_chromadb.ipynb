```
from sentence_transformers import SentenceTransformer
import chromadb
from chromadb.config import Settings
from chromadb.utils.embedding_functions import SentenceTransformerEmbeddingFunction

client = chromadb.Client(Settings(chroma_db_impl="duckdb+parquet", persist_directory="/chroma/chroma"))
collection = client.create_collection("docs", embedding_function=SentenceTransformerEmbeddingFunction())

docs = ["Transformers are a deep learning model.", "LLaMA is a large language model developed by Meta."]
collection.add(documents=docs, ids=["1", "2"])

query = "What are transformers?"
results = collection.query(query_texts=[query], n_results=1)
print("Top document:", results["documents"][0][0])
```
